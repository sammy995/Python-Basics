#Matplotlib is a Python library for creating static, animated, and interactive visualizations in Python.

Matplotlib can be used in Python scripts, the Python and IPython shells, the Jupyter notebook, web application servers, and four graphical user interface toolkits.

Matplotlib can be used to plot data from CSV files, arrays, and other data structures.

Matplotlib supports a variety of plot types, including line plots, scatter plots, bar plots, and histograms.

Matplotlib also supports a variety of customization options, such as changing the plot title, axis labels, and line colors.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot')
```

```python
plt.plot(a,b,'ro--',  markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Line Graph')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'go')
plt.subplot(2,2,2)
plt.plot(x,a,'b*')
plt.subplot(2,2,4)
plt.plot(a,b,'r--')
```

```python
plt.bar(x,a,color='g')
```

```python
a=np.array([12,4,2,1,414,5,51,6,16,7,2,4,56,8,9,0,3,1,3,4,56,89,3,1,3,3,3,33,4,5,6,677,8,9,4,2,1])
plt.hist(a,bins=70)
```

```python
data=[np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.boxplot(data,vert=True,patch_artist=True)
```

#Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn is designed to work well with Pandas data structures and is tightly integrated with the PyData stack.

Some of the key features of Seaborn include:

* A simple and concise API that makes it easy to create visualizations.
* A wide range of built-in plot types, including line plots, scatter plots, bar plots, histograms, and box plots.
* Extensive support for customization, allowing you to create visualizations that are tailored to your specific needs.
* Integration with Pandas data structures, making it easy to create visualizations from your data.

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```
