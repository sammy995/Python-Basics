```python
# prompt: Definition of List and some examples

# Definition of a list
# A list is a collection of items in a particular order.
# Each item in a list is called an element.
# Lists are ordered collections, meaning that the items have a defined order, and that order will not change.
# The items in a list can be of any data type, and they can be different data types.

# Examples of lists
my_list = []  # Empty list
my_list = [1, 2, 3]  # List of integers
my_list = ["apple", "banana", "cherry"]  # List of strings
my_list = [True, False, False]  # List of booleans
my_list = [1, "apple", True]  # List of different data types

# Accessing elements in a list
# You can access the elements of a list using the index operator [].
# The index operator takes an integer as an argument, and it returns the element at that index.
# The first element of a list has an index of 0, the second element has an index of 1, and so on.

# Example of accessing elements in a list
print(my_list[0])  # Output: 1
print(my_list[1])  # Output: "apple"
print(my_list[2])  # Output: True

# Modifying elements in a list
# You can modify the elements of a list using the assignment operator =.
# The assignment operator takes two arguments: the index of the element to modify, and the new value for the element.

# Example of modifying elements in a list
my_list[0] = 5
print(my_list)  # Output: [5, "apple", True]

# Adding elements to a list
# You can add elements to a list using the append() method.
# The append() method takes one argument: the element to add to the list.
# The element will be added to the end of the list.

# Example of adding elements to a list
my_list.append("orange")
print(my_list)  # Output: [5, "apple", True, "orange"]

# Removing elements from a list
# You can remove elements from a list using the remove() method.
# The remove() method takes one argument: the element to remove from the list.
# The element will be removed from the first occurrence of the element in the list.

# Example of removing elements from a list
my_list.remove("orange")
print(my_list)  # Output: [5, "apple", True]

# The del statement
# The del statement can be used to delete elements from a list.
# The del statement takes one argument: the index of the element to delete.

# Example of using the del statement
del my_list[1]
print(my_list)  # Output: [5, True]
```

#Tuple


A tuple is an ordered sequence of immutable Python objects. Tuples are sequences, just like lists. The differences between tuples and lists are:

* Tuples are immutable. This means that you cannot add, remove, or modify items in a tuple.
* Tuples are faster than lists. This is because tuples are stored in a more compact way than lists.

Tuples are created using parentheses, (), and the items in a tuple are separated by commas.

```python
my_tuple=tuple()
```

```python
type(my_tuple)
```

```python
my_tuple=("Audi","Merc","BMW")
```

```python
my_tuple.count("Audi")
```

```python
my_tuple.index('BMW')
```
